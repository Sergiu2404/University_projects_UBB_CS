```python
# 2.
A = matrix([[1, 1.0/2], [0, 0.6]])
A2 = A * A
A3 = A2 * A
print(A2)
print()
print(A3)
```

```python
def compute_matrix_exponentiation(A, n):
    product = A
    for i in range(n - 1):
        product *= A
        
    return product
```

```python
print(compute_matrix_exponentiation(A, 2))
print()
print(compute_matrix_exponentiation(A, 3))
```

```python
for i in range(10):
    exponent = 1000000
    print(compute_matrix_exponentiation(A, exponent))
    print()
    exponent += 1000000
```

```python
# 1.
t = var('t')
y = function('y')(t)

deq = diff(y, t) == 1 + y^2
exact_solution = desolve(deq, y, ics=[0, 1])
exact_solution = tan(exact_solution.rhs())
print(exact_solution)

plot(exact_solution, (t, -1, 0.7), ymin=-1, ymax=2.5)
```

```python
n = 100
def f(x):
    return 1 + x^2

def approx_point(x):
    xk = 0
    yk = 1
    h = x / n # Step size
    
    for i in range(n):
        yk = yk + h * f(xk)
        xk += h
        
    return yk
```

```python
print(approx_point(0.5))
```

```python
plot(approx_point, (x, -2, 2), ymin=-2, ymax=3)
```
