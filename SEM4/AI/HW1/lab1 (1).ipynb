```python
#HW lab 1
#TASK 6: a) Consider an unfair probability spinner with n slots. Each slot has a probab
# Pn in [0, 1] and SUM(Pi) = 1, i = 0,n-1; Write a function that random generate an 
# index i with the probab Pi from {0, ..., n}.
# b) The same pb but instead of a distribution the input list is a list of natural nr that holds
# degrees number that the center angle Theta(i) correspondent to index i has.
# (instead of 1 -> 360)
import numpy as np
import matplotlib.pyplot as plt

def spinner(probabilityDistribution):
    random_number = np.random.rand()
    cumulative_probability = 0
    i=0

    while i < len(probabilityDistribution):
        cumulative_probability += probabilityDistribution[i]
        if random_number <= cumulative_probability:
            return i
        i += 1

    

def spinner_outcomes(probabilityDistribution, n_trials):
    outcomes = [spinner(probabilityDistribution) for _ in range(n_trials)]

    plt.hist(outcomes, bins=len(set(outcomes)), align='mid', rwidth=0.5)
    plt.xlabel('Index')
    plt.ylabel('Frequency of indexed slot')
    plt.xticks(range(1, len(set(outcomes)) + 1))
    plt.show()


probabilityDistribution = [0.2, 0.3, 0.1, 0.1, 0.3]

n_trials = 10000
spinner_outcomes(probabilityDistribution, n_trials)

slot = spinner(probabilityDistribution)
print(f"Single outcome: The spinner is on slot {slot}")





# b) The same pb but instead of a distribution the input list is a list of natural nr that holds
# degrees number that the center angle Theta(i) correspondent to index i has.
# (instead of 1 -> 360)

def spinner_angles_outcomes(probabilityDistribution, n_trials):
    outcome = [spinner_angles(probabilityDistribution) for _ in range(n_trials)]

    plt.hist(outcome, bins=len(set(outcome)), align='mid', rwidth=0.5)
    plt.xlabel('Index')
    plt.ylabel('Frequency of indexed slot')
    plt.xticks(range(1, len(set(outcome)) + 1))
    plt.show()

def spinner_angles(angleProbabilities):
    
    random_number = np.random.rand()
    cumulative_probability = 0
    
    # i=0
    # while i < len(angleProbabilities):
    #     cumulative_probability += probabilityDistribution[i]
    #     if random_number <= cumulative_probability:
    #         return i
    #     i += 1

    for i, Pi in enumerate(angleProbabilities):
        cumulative_probability += Pi
        if random_number <= cumulative_probability:
            return i

angles_list = np.array([160,85,45,70])
angleProbabilities = angles_list/360
print(angleProbabilities)

n_trials = 10000
spinner_angles_outcomes(angleProbabilities, n_trials)

slot = spinner_angles(angleProbabilities)
print(f"Single outcome: The spinner is on slot {slot}")
```
